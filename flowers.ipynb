```python
import pandas as pd
import numpy as np
```

```python
og_df = pd.read_csv("flowers.csv")
cleaned = og_df
```

```python
og_df.head()
```

```python
cleaned[['height (min)', 'height (max)']] = cleaned['height (cm)'].str.split('-', expand=True)
cleaned[['longevity (min)', 'longevity (max)']] = cleaned['longevity (years)'].str.split('-', expand=True)
cleaned[['petals (min)', 'petals (max)']] = cleaned['average number of petals'].str.split('-', expand=True)

cleaned = cleaned.drop(columns=["average number of petals", "height (cm)", "longevity (years)"])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
cleaned["climate (encoded)"] = le.fit_transform(cleaned["climate"])
cleaned["place of origin (encoded)"] = le.fit_transform(cleaned["place of origin"])

cleaned.head()
```

```python
cleaned[['color_1', 'color_2', 'color_3']]= cleaned['color'].str.split(', ', expand=True)

for col in ['color_1', 'color_2', 'color_3']:
    cleaned[col] = cleaned[col].str.lower()

for col in ['color_1', 'color_2', 'color_3']:
    cleaned[col] = le.fit_transform(cleaned[col])

cleaned = cleaned.drop(columns=["climate", "place of origin", "color"])
cleaned.head()
```
